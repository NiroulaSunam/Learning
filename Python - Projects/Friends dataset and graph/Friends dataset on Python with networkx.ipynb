```python
df_1994 = df[df.values == 1994]
```

```python
df_1994
```

```python
G1 = nx.from_pandas_edgelist(df_1994, source='Year_of_prod', target='Episode_Title', edge_attr=True)
G2 = nx.from_pandas_edgelist(df_1994, source='Year_of_prod', target='Summary', edge_attr=True)
G3 = nx.compose(G1,G2)
```

```python
nx.draw_networkx(G3)
plt.show()
```

```python
nx.write_gexf(G3, "friends_1994_graph.gexf")
```

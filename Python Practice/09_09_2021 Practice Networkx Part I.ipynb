```python
import networkx as nx
```

```python
G = nx.Graph()
```

```python
G.add_node(1)
```

```python
G.add_node(2)
```

```python
G.add_node(3)
```

```python
G.add_edge(1,2)
G.add_edge(2,3)
G.add_edge(1,3)
```

```python
print(G)
```

```python
print(G.nodes)
```

```python
print(G.edges)
```

```python
import matplotlib.pyplot as plt
```

```python
nx.draw(G)
plt.show()
```

```python
G1 = nx.Graph()
```

```python
l = [1,2,3,4,5]
```

```python
G1.add_nodes_from(l)
```

```python
G1.add_edge(1,2)
G1.add_edge(2,3)
G1.add_edge(1,3)
G1.add_edge(1,5)
G1.add_edge(2,4)
```

```python
print(G1)
```

```python
print(G1.nodes)
```

```python
print(G1.edges)
```

```python
nx.draw(G1)
plt.show()
```

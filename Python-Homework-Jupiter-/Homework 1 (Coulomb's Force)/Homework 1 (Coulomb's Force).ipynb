## Electrostatic force between the charges
### The Coulomb's force between two charges $q_1$ and $q_2$ separated at distance $r$ is given as; 
#### $ F = k \frac{q_1q_2}{r^2}$, where $ k = 9\times 10^9 Nm^2C^{-2}$

- Write a code for function to calculate the Coulomb's force between two given charge at given distance
- use list to calculate the variation of Coulomb's force between two charges (as you like) placed at a range of distance (choose range in your favour)
- Make a plot that shows the variation of Coulomb's force with distance using list

#### A code for function to calculate the Coulomb's force between two given charge at given distance

```python
def CoulombForce(q1,q2,r):
    F = ((9*(10**9))*q1*q2)/(r*r)
    return F
```

```python
a = CoulombForce(1,2,3)
print(a)
```

#### A list to calculate the variation of Coulomb's force between two charges (already given) at a range of distance (range of my choosing)

```python
def CoulombForce2(r):
    q1 = 2
    q2 = 3
    f = ((9*(10**9))*q1*q2)/(r*r)
    return f
```

```python
R = []
F = []
for i in range (1,10):
    R.append(i)
    f1 = CoulombForce2(i)
    F.append(f1)
print(R)
print(F)
```

#### Plot showng the variation of Columb's Force and the distance using the list

```python
import matplotlib.pyplot as plt
plt.plot(F,R)
plt.xlabel("Force 'F'",fontsize = 15)
plt.ylabel("Distance 'd'",fontsize = 15)
plt.title("Coulomb's Force w.r.t. increasing distance",fontsize = 15)
plt.scatter(F,R,color = 'r')
plt.savefig("Coulomb's Force w.r.t increasing distance.jpg")
plt.savefig("Coulomb's Force w.r.t increasing distance.pdf")
```

### Calculation of F using a different number of data sets (q1,q2,r) and variable data and its plotting

```python
def Coulumbforce():   
    """ Enter the total number of data you have to calculate. And You will then be asked to enter q1,q2,r 
    of Columb's Force and Force is printed and Force and Radius are returned """
    F = []
    R = []
    a = int(input("Enter the total number of data you have: "))
    for i in range (0,a):
        q1 = float(input("Enter charge q1: "))
        q2 = float(input("Enter charge q2: "))
        r = float(input("Enter the distance between charge q1 and q2; r: "))
        k = 9*(10**9)
        f = (k*q1*q2)/(r*r)
        F.append(f)
        R.append(r)
    print("Columb Force, F =", F)
    return (F,R)
```

```python
b = Coulumbforce()
```

```python
import matplotlib.pyplot as plt
plt.plot(b[0],b[1])
plt.xlabel('Force "F"', fontsize = 15)
plt.ylabel("Distance 'd'", fontsize = 15)
plt.title("Columb Force",fontsize =20)
plt.savefig('Columb Force (Plot between Force and Distance).jpg')
plt.savefig('Columb Force (Plot between Force and Distance).pdf')
```

### Plot using Random Variables of q1,q2,r (It will give different points on the graph everytime it runs)

```python
import numpy as np
import matplotlib.pyplot as plt
k = 9 * (10**9)
A = np.arange(1,10)
q1 = np.random.permutation(A)
q2 = np.random.permutation(A)
r = np.random.permutation(A)
F = k*((q1 * q2))/(r * r)
plt.plot(F,r,'-.k')
plt.scatter(F,r,color = 'r')
plt.xlabel('Force "F"', fontsize = 15)
plt.ylabel("Distance 'd'", fontsize = 15)
plt.title("Columb Force",fontsize =20)
plt.savefig("Random Coulumb Force.jpg")
plt.savefig("Random COulumb Force.pdf")
```

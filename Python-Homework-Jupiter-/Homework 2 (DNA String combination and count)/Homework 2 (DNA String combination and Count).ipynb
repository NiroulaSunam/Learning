### Create a DNA string of 10,000 characters and count the following: A,T,G,C,all combination of two charaters, all combinations of three characters

```python
import random as rn
D = []
dna = ['A','T','G','C']
for i in range (10000):
    a = rn.choice(dna)
    D.append(a)
DNA = ''.join(D)
print(DNA)
```

```python
A = {}
for i in dna:
    a = DNA.count(i)
    print(i,'=',a)
    A[i]=a
print(A)
```

```python
J = []
B = {}
from itertools import combinations
comb = list(combinations(dna,2))
for i in comb:
    j = ''.join(i)
    J.append(j)
for j in J:
    b = DNA.count(j)
    print(j,'=',b)
    B[j]=b
print(B)
```

```python
J = []
C = {}
from itertools import combinations
comb = list(combinations(dna,3))
for i in comb:
    j = ''.join(i)
    J.append(j)
for j in J:
    c = DNA.count(j)
    print(j,'=',c)
    C[j]=c
print(C)
```

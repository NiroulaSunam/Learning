# Task 1: Create Class: Gravitation
### The garavitational force, potential and intensity are given as;
- Gravitational force between two masses sepateted by some distance
- $F = G \frac{m_1m_2}{r^2}$
- Gravitational potential of a mass at some distance
- $V = G \frac{m_1}{r}$
- Gravitational intensity of a mass at some distance
- $g = G \frac{m_1}{r^2}$
### Create a class of these three function

```python
class Gravitation1():
    def __init__ (self,G,m1,m2,r):
        self.G = G
        self.m1 = m1
        self.m2 = m2
        self.r = r
        
    def gforce(self):
        """To produce Gravitational force in G,m1,m2,r
        G.gforce(r)"""
        F = (self.G*self.m1*self.m2)/(self.r**2)
        return F
    
    def gpotential(self):
        """To produce Gravitational potential in G,m1,m2,r 
        gpotential(G,m1,m2,r) """
        P = (self.G*self.m1)/(self.r)
        return P
    
    def gintensity(self):
        """To produce Gravitational intensity in G,m1,m2,r 
        gintensity(G,m1,m2,r)"""
        I = (self.G*self.m1)/(self.r**2)
        return I
```

```python
G1 = Gravitation1(1,2,3,4)
```

```python
G1.gforce()
```

```python
G1.gpotential()
```

```python
G1.gintensity()
```

### Use lists to calculate functional value with variation of distance r within some range

```python
class Gravitation2():
    def __init__ (self,G,m1,m2):
        self.G = G
        self.m1 = m1
        self.m2 = m2
        
    def gforce(self,r):
        """To produce Gravitational force in G,m1,m2,r
        G= Gravitation(G,m1,m2,r) G.gforce(r)"""
        F = (self.G*self.m1*self.m2)/(r**2)
        return F
    
    def gpotential(self,r):
        """To produce Gravitational potential in G,m1,m2,r 
        G= Gravitation(G,m1,m2,r) G.gpotential(r) """
        P = (self.G*self.m1)/(r)
        return P
    
    def gintensity(self,r):
        I = (self.G*self.m1)/(r**2)
        return I
```

```python
G2 = Gravitation2(1,2,3)
```

```python
G2.gforce(4)
```

```python
G2.gpotential(4)
```

```python
G2.gintensity(4)
```

```python
G = Gravitation2(6.67e-11,1,2)
R = []
F = []
P = []
I = []
for r in range (1,1000):
    f = G.gforce(r)
    p = G.gpotential(r)
    i = G.gintensity(r)
    R.append(r)
    F.append(f)
    P.append(p)
    I.append(i)
```

### Use matplotlib to plot the variation of function with distance r
### Try to plot semilog and log log plot also

```python
import matplotlib.pyplot as plt

plt.figure(figsize = [12,12])
plt.subplot(2,2,1)
plt.plot(R,F, color = 'b', label = 'Force')
plt.plot(R,P, color = 'k', label = 'Potential')
plt.plot(R,I, color = 'r', label = 'Intentity')
plt.legend()
plt.xlabel('R', fontsize = 12)
plt.ylabel('F,V,g', fontsize = 12)
plt.title('F,V,g vs R graph', fontsize = 15)
plt.savefig("Normal Force,Potential,Radius graph.jpeg")
plt.savefig("Normal Force,Potential,Radius graph.pdf")

plt.subplot(2,2,2)
plt.semilogy(R,F, color = 'b', label = 'Force')
plt.semilogy(R,P, color = 'k', label = 'Potential')
plt.semilogy(R,I, color = 'r', label = 'Intentity')
plt.legend()
plt.xlabel('R', fontsize = 12)
plt.ylabel('F,V,g', fontsize = 12)
plt.title('F,V,g vs R graph', fontsize = 15)
plt.savefig("Semilog Force,Potential,Radius graph.jpeg")
plt.savefig("Semilog Force,Potential,Radius graph.pdf")

plt.subplot(2,2,3)
plt.loglog(R,F, color = 'b', label = 'Force')
plt.loglog(R,P, color = 'k', label = 'Potential')
plt.loglog(R,I, color = 'r', label = 'Intentity')
plt.legend()
plt.xlabel('R', fontsize = 12)
plt.ylabel('F,V,g', fontsize = 12)
plt.title('F,V,g vs R graph', fontsize = 15)
plt.savefig("Log Force,Potential,Radius graph.jpeg")
plt.savefig("Log Force,Potential,Radius graph.pdf")
```

# Task 2: Electrostatic vs Gravitational force
### Use following data for calculation;
- mass of electron and proton; $9.1\times10^{-31}kg, 1.67 \times 10^{-27} kg$
- charge of electron and proton; $ 1.6 \times 10^{-19}C$
- Distance between electron and proton in H- atom; $5.29\times 10^{-11}m$
### write the suitable code for electrostatic force between two charges and gravitational force between two masses
### Calculate the electrostatic force and gravitational force between electron and proton within H - atom and compare them
    

```python
class Comparision():
    """To Calculate the gravitational force and Electrostatic force Comparision(G,K,m1,m2,q1,q2,r)"""
    
    def __init__(self,G,K,m1,m2,q1,q2,r):
        self.G = G
        self.K = K
        self.m1 = m1
        self.m2 = m2
        self.q1 = q1
        self.q2 = q2
        self.r = r
        
    def gforce(self):
        F1 = (self.G*self.m1*self.m2)/(self.r**2)
        return F1
    
    def eforce(self):
        F2 = (self.K*self.q1*self.q2)/self.r**2
        return F2
```

```python
C = Comparision(6.67e-11,9e9,9.1e-31,1.67e-27,1.6e-19,1.6e-19,5.29e-11)
g = C.gforce()
e = C.eforce()

if g > e:
    print("Gravitational force is greater than Electrostatic force ")
else:
    print("Electrostatic force is greater than Gravitational force ")
```

### Using the lists calculate the electrostatic and gravitational force between electron and protron for a range of separation

```python
class Comparision2():
    
    def __init__(self,G,K,m1,m2,q1,q2):
        self.G = G
        self.K = K
        self.m1 = m1
        self.m2 = m2
        self.q1 = q1
        self.q2 = q2
        
    def gforce(self,r):
        F1 = (self.G*self.m1*self.m2)/(r**2)
        return F1
    
    def eforce(self,r):
        F2 = (self.K*self.q1*self.q2)/(r**2)
        return F2
```

```python
C2 = Comparision2(6.67e-11,9e9,9.1e-31,1.67e-27,1.6e-19,1.6e-19)
R = []
GF = []
EF = []
for r in range (1,1000):
    gf = C2.gforce(r)
    ef = C2.eforce(r)
    R.append(r)
    GF.append(gf)
    EF.append(ef)
```

### Plot the variation of electrostatic and gravitational force between them using matplotlib
### Use semilog and loglog plot for the comparison of the electrostatic and gravitational force.

```python
import matplotlib.pyplot as plt
plt.figure(figsize = [13,13])
plt.subplot(2,2,1)
plt.plot(R,GF, label = "Gravitational force")
plt.plot(R,EF, label = "Electrostatic force")
plt.ylabel("Gravitational and Electrostatic force")
plt.xlabel("Radius")
plt.title("Compaing Gravit. and Elect. force")
plt.legend()

plt.subplot(2,2,2)
plt.semilogy(R,GF, label = "Gravitational force")
plt.semilogy(R,EF, label = "Electrostatic force")
plt.ylabel("Gravitational and Electrostatic force")
plt.xlabel("Radius")
plt.title("Compaing Gravitational and Electrostatic force")
plt.legend()

plt.subplot(2,2,3)
plt.loglog(R,GF, label = "Gravitational force")
plt.loglog(R,EF, label = "Electrostatic force")
plt.ylabel("Gravitational and Electrostatic force")
plt.xlabel("Radius")
plt.title("Compaing Gravitational and Electrostatic force")
plt.legend()
plt.savefig("Gravitational and Electrostatic force comparision.jpeg")
plt.savefig("Gravitational and Electrostatic force comparision.pdf")
```

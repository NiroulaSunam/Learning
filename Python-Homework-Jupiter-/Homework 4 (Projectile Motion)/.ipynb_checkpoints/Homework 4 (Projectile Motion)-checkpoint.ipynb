# Projectile Motion 
### Range
- $R = \frac{u^2 sin2\theta}{2g}$

### Time of Flight 
- $T = \frac{2u sin\theta}{g}$

### Horizontal Range
- $H = \frac{u^2 sin^2\theta}{2g}$

## Task:1  Make a class to calculate the range, time of flight and horizontal range of the projectile fired from the ground.

u, $\theta$, $sin2\theta$, g, degree to radian - multiply by 0.01745329

```python
import math
class Projectile():
    ''' To Determine the Range, Time of Flight(TOF) and Horizontal Range(HR) of Projectile fired from the ground. 
    Projectile(u,th,g)
    where, u = initial velocity
           th = angle made or theta
           g = acceleration due to gravity'''
    def __init__(self,u,th,g):
        self.u = u
        self.th = th * 0.01745329
        self.g = g
    
    def Range(self):
        R = (self.u**2)*(math.sin(2*(self.th)))/(2*self.g)
        return R
    
    def TOF(self):
        TOF = (2*self.u*(math.sin(self.th)))/self.g 
        return TOF
    
    def HR(self):
        HR = (self.u**2)*(math.sin(self.th)**2)/2*self.g
        return HR
    
# help(Projectile)
```

```python
P = Projectile(12,27.5,9.8)
```

```python
print(P.Range())
print(P.TOF())
print(P.HR())
```

## Task: 2 Use the list to find the range, time of flight and horizontal range for variying value of angle from 1 degree to 90 degree

```python
import math
class Projectile2():
    ''' To Determine the Range, Time of Flight(TOF) and Horizontal Range(HR) of Projectile fired from the ground. 
    Projectile(u,g) at variable angle
    where, u = initial velocity
           g = acceleration due to gravity'''
    def __init__(self,u,g):
        self.u = u
        self.g = g
    
    def Range(self,th):
        R = (self.u**2)*(math.sin(2*(th*0.01745329)))/(2*self.g)
        return R
    
    def TOF(self,th):
        TOF = (2*self.u*(math.sin(th*0.01745329)))/self.g 
        return TOF
    
    def HR(self,th):
        HR = (self.u**2)*(math.sin(th*0.01745329)**2)/2*self.g
        return HR
```

```python
P = Projectile2(20,9.8)
Tt = []
Rg = []
Tf = []
Hr = []
for i in range(1,91):
    rg = P.Range(i)
    tf = P.TOF(i)
    hr = P.HR(i)
    Tt.append(i)
    Rg.append(rg)
    Tf.append(tf)
    Hr.append(hr)
```

## Task:3 Make a plot to show the variation of range, time of flight and horizontan range with angle of projection.

```python
import matplotlib.pyplot as plt
plt.figure (figsize = [12,12])
plt.subplot(2,2,1)
plt.plot(Tt,Rg)
plt.xlabel('Angle', fontsize = 15)
plt.ylabel('Range',fontsize = 15)
plt.title('Angle vs Range', fontsize = 15)
plt.suptitle('Var. of Range, Time of Flight, and Horizontal Range with Angle of Proj.',fontsize=20)

plt.subplot(2,2,2)
plt.plot(Tt,Tf)
plt.xlabel('Angle', fontsize = 15)
plt.ylabel('Time of Flight',fontsize = 15)
plt.title('Angle vs Time of Flight', fontsize = 15)

plt.subplot(2,2,3)
plt.plot(Tt,Hr)
plt.xlabel('Angle', fontsize = 15)
plt.ylabel('Horizontal Range',fontsize = 15)
plt.title('Angle vs Horizontal Range', fontsize = 15)
plt.savefig('Variation of Range, Time of Flight and Horizontal Range with Angle of Projection.jpeg')
plt.savefig('Variation of Range, Time of Flight and Horizontal Range with Angle of Projection.pdf')
```

```python
plt.figure (figsize = [12,12])
plt.subplot(2,2,1)
plt.plot(Tt,Rg, label = 'Range')
plt.plot(Tt,Tf, label = 'Time of Flight')
plt.plot(Tt,Hr, label = 'Horizontal Range')
plt.xlabel('Angle', fontsize = 12)
plt.ylabel('Range, Time of Flight, Horizontal Range', fontsize = 12)
plt.title('Angle vs R,T,H', fontsize = 15)

plt.subplot(2,2,2)
plt.plot(Tt,Rg, label = 'Range')
plt.plot(Tt,Tf, label = 'Time of Flight')
plt.semilogy(Tt,Hr, label = 'Horizontal Range')
plt.xlabel('Angle', fontsize = 12)
plt.ylabel('Range, Time of Flight, Horizontal Range', fontsize = 12)
plt.title('Angle vs R,T,logH', fontsize = 15)
plt.legend()

plt.subplot(2,2,3)
plt.semilogy(Tt,Rg, label = 'Range')
plt.semilogy(Tt,Tf, label = 'Time of Flight')
plt.semilogy(Tt,Hr, label = 'Horizontal Range')
plt.xlabel('Angle', fontsize = 12)
plt.ylabel('Range, Time of Flight, Horizontal Range', fontsize = 12)
plt.title('Angle vs logR,logT,logH', fontsize = 15)
plt.legend()
plt.savefig('Variation of Range, Time of Flight and Horizontal Range and their log with Angle of Projection.jpeg')
plt.savefig('Variation of Range, Time of Flight and Horizontal Range and their log with Angle of Projection.pdf')
```

## Task: 4 Change the list of [angle], [Range], [Time of flight], and [Horizontal range] into dictionary and finally into a DataFrame using pandas. Save the file in your PC in csv file. 

```python
Data = {}
Data.update({'Angle':Tt,'Range':Rg,'Time of Flight':Tf,'Horizontal Range':Hr})
import pandas as pd
SP = pd.DataFrame(Data)
print(SP)
```

```python
SP.to_csv('Projectile Motion.csv')
```
